```python
import numpy as np
from numpy import linalg as LA
import matplotlib.pyplot as plt
```

```python
a=np.array([[0.025,-0.01,0.011],
            [-0.01,0.035,-0.0095],
            [0.011,-0.0095,0.1]])
b=np.array([[1,1,-1],
            [-2,1,0],
            [1,1,1]])
c=np.array([[3,-1,-5],
            [-1,2,3],
            [-5,3,10]])
```

```python
def AnisotropyTensor(x):
    k=(np.trace(x)/2)
    a=x/(2*k)-((1/3)*np.eye(3))
    alambda,aEpsilon = LA.eig(a)
    II=alambda[0]*alambda[0]+alambda[1]*alambda[1]+alambda[0]*alambda[1]
    III=-alambda[0]*alambda[1]*(alambda[0]+alambda[1])
    epsilon=np.power(III/2,1/3)
    eta=np.power(II/3,1/2)
    xb=alambda[0]-alambda[1]+(3/2)*alambda[2]+(1/2)
    yb=(np.power(3,1/2)/2)*(3*alambda[2]+1)
    return a,k,alambda,aEpsilon,III,II,epsilon,eta,xb,yb
```

```python
AnisotropyTensor(a)
```

```python
# Eigenvalue Map
fig = plt.figure()
ax = fig.add_subplot(111)
x1 = np.linspace(-1/3,1/6,100)
x2 = np.linspace(-1/3,0,100)
x3 = np.linspace(0,1/6,100)
y1=-1.*x1+(1/3)
y2=-2.*x2
y3 = x3
plt.plot(x1,y1)
plt.plot(x2,y2)
plt.plot(x3,y3)
plt.plot(AnisotropyTensor(a)[2][1],AnisotropyTensor(a)[2][0],'co')
ax.annotate('X1c', xy=(-1/3, 2/3), xytext=(-0.37, 0.65))
ax.annotate('X2c', xy=(1/6, 1/6), xytext=(0.17, 0.17))
ax.annotate('X3c', xy=(0, 0), xytext=(-0.01, -0.03))
plt.xlabel('$\lambda_{2}$',fontsize=15)
plt.ylabel('$\lambda_{1}$',fontsize=15)
ax.set_xlim(-0.4,0.25)
ax.set_ylim(-0.05,0.7)
plt.show()
```
